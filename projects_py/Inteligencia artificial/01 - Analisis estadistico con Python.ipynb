**Ejemplo de graficas en una presentacion interactiva**

$$\text{Sea } 0 \leqslant x \leqslant 10$$
$$f_1(x) = x$$
$$f_2(x) = x^2$$
$$f_3(x) = x^3$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Rango de valores para x
# np.arange(minimo, maximo, paso)
x = np.arange(0, 10, 0.4)
```

```python
x
```

```python
plt.plot(x, x, 'g--')
```

```python
y1 = x
y2 = x**2
y3 = x**3
```

```python
plt.plot(x, y1, 'g--', x, y2, 'bs', x, y3, 'r^')
plt.show()
```
