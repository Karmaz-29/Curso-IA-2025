```python
import cv2
import common
import pylab
import cv2
import pytesseract
import matplotlib.pyplot as plt
#Para windows
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
```

```python
placa = []
ej01 = cv2.imread('../data_set/JPG/ej01-auto.jpg')
gray1 = cv2.cvtColor(ej01, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(cv2.cvtColor(ej01, cv2.COLOR_BGR2RGB));
```

```python
count = 0
while count <= 2:
    gray1 = cv2.blur(gray1, (3,3))
    canny1 = cv2.Canny(gray1, 130,200)
    canny1 = cv2.dilate(canny1, None, iterations=1)
    count += 1
plt.imshow(canny1, cmap='gray')
```

```python
contorno1, _ = cv2.findContours(canny1, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
```

```python
for c in contorno1:
    area = cv2.contourArea(c)
    x, y, w, h = cv2.boundingRect(c)
    epsilon = 0.9 * cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, epsilon, True)
    if area >= 5000:
        relAspec = float(w)/h
        cv2.drawContours(ej01, [approx], 0, (0,255,0), 3)
        print(f'Área: {area}, Coordenadas: x={x}, y={y}, w={w}, h={h}, Relacion de aspecto={relAspec}')
    
        if relAspec > 1:
            placa = gray1[y:y+h, x:x+w]
            text = pytesseract.image_to_string(placa, config='--psm 11'.strip())
            print('Placa: ', text)
            cv2.rectangle(ej01, (x, y), (x+w, y+h), (0,255,0), 3)
            cv2.putText(ej01, text, (x-20, y-20), 1, 2.2, (0,255,0), 3)
```

```python
plt.imshow(cv2.cvtColor(ej01, cv2.COLOR_BGR2RGB));
```
