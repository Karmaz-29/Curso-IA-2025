```python
from matplotlib import pyplot as plt
import numpy as np
import imageio.v2 as imageio
from skimage import color

import cv2
import common
import pylab
```

```python
# modificar las dimensiones de visualizacion
pylab.rcParams['figure.figsize'] = (6.0,4.0)
ruta = '../data_set/JPG/wall-e.jpg'
input_image = cv2.imread(ruta)
```

```python
plt.imshow(input_image)
```

```python
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB));
```

```python
flipped_img = cv2.flip(input_image, 0)
plt.imshow(cv2.cvtColor(flipped_img, cv2.COLOR_BGR2RGB));
```

```python
flipped_img = cv2.flip(input_image, 1)
plt.imshow(cv2.cvtColor(flipped_img, cv2.COLOR_BGR2RGB));
```

```python
transpuesta = cv2.transpose(input_image)
plt.imshow(cv2.cvtColor(transpuesta, cv2.COLOR_BGR2RGB));
```

```python
plt.imshow(cv2.cvtColor(cv2.flip(transpuesta, 1), cv2.COLOR_BGR2RGB));
```

```python
blank_image = np.zeros((input_image.shape), np.uint8)
blank_image[125:255, 350:450, 1] = 50
plt.imshow(blank_image)
```

```python
nueva = cv2.cvtColor(cv2.add(blank_image, input_image), cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(nueva);
```

```python
# Filtro de desenfoque
d = 3
img_desenfoque = cv2.GaussianBlur(input_image, (2*d+1, 2*d+1), - 1)[d:-3, d:-d]
plt.imshow(cv2.cvtColor(img_desenfoque, cv2.COLOR_BGR2RGB));
```

```python
grayImage = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
plt.imshow(grayImage, cmap='gray');
```

```python
sobelX = cv2.Sobel(grayImage, cv2.CV_64F, 1, 0, ksize=9)
plt.imshow(sobelX, cmap='gray');
```

```python
sobelX = cv2.Sobel(grayImage, cv2.CV_64F, 0, 1, ksize=9)
plt.imshow(sobelX, cmap='gray');
```

```python
sobelX = cv2.Sobel(grayImage, cv2.CV_64F, 1, 1, ksize=9)
plt.imshow(sobelX, cmap='gray');
```

## Maximos y minimos

```python
for i in range(0,3):
    minimo, maximo, min_loc, max_loc = cv2.minMaxLoc(input_image[:, :, i])
    print('min {} esta ubicado en {} y el maX {} esta ubicado en {}'.format(minimo, maximo, min_loc, max_loc))
```
