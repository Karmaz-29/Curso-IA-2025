```python
# importar paquetes
from matplotlib import pyplot as plt
import numpy as np
import imageio.v2 as imageio
from skimage import color
```

```python
ruta = '../data_set/JPG/gray-wolf.jpg'
```

```python
imgIn = imageio.imread(ruta)
```

```python
# Caracteristicas de la imagen = matriz h * w * 3
print('Dimensiones de la imagen: ', imgIn.shape)
```

```python
# Imprimir contenido de las capas de color
print('Pixel en [0,0,0]', imgIn[0,0,0])
print('Pixel en [0,0,1]', imgIn[0,0,1])
print('Pixel en [0,0,2]', imgIn[0,0,2])
print('Pixel en [0,0,:]', imgIn[0,0,:])
```

```python
print('Tipo de datos: ', imgIn.dtype)
```

```python
print('Valor maximo', imgIn.max())
print('Valor minimo', imgIn.min())
```

```python
plt.imshow(imgIn);
```

```python
imgGray = color.rgb2gray(imgIn)
print('Dimensiones en la escala de grises: ', imgGray.shape)
print('Tipo de dato contenido: ', imgGray.dtype)
```

```python
plt.imshow(imgGray);
```

```python
plt.imshow(imgGray, cmap='gray');
```

```python
imgSeccion = imgGray[110:440, 215:525]
plt.imshow(imgSeccion, cmap='gray')
```

```python
imgModificada = imgGray.copy()
imgModificada[imgModificada < 0.4 ] = 0.0
plt.imshow(imgModificada, cmap='gray')
```

```python
plt.hist(imgGray.flatten());
```

```python
plt.hist(imgGray.flatten(),bins = np.linspace(0,1,256));
```
