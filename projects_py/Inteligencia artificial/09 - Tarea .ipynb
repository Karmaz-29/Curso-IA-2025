```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
ruta = '../data_set/CSV/Mall_Customers-2.csv'
```

```python
df = pd.read_csv(ruta, index_col=0)
```

```python
df
```

```python
df.info()
```

```python
df.rename({'Gender': 'Genero', 'Age':'Edad', 'Annual Income (k$)':'Ingreso', 'Spending Score (1-100)':'Gasto'}, axis=1, inplace=True)
```

```python
df
```

```python
df.describe().T
```

```python
df['Segmento'] = np.where(df.Ingreso >= 90, 'Ingreso alto', np.where(df.Ingreso < 50, 'Ingreso bajo', 'Ingreso moderado'))
```

```python
df
```

```python
train, test = train_test_split(df, test_size=0.4, stratify=df['Segmento'], random_state=23)
```

```python
X_train = train[['Edad', 'Ingreso', 'Gasto']]
Y_train = train['Segmento']

# Conjunto de pruebas
X_test = test[['Edad', 'Ingreso', 'Gasto']]
Y_test = test['Segmento']
```

```python
mod_gnb = GaussianNB()
y_pred = mod_gnb.fit(X_train, Y_train).predict(X_test)
print('La prediccion del CLasificador Gaussiano Navie Bayes en los datos de prueba es {:0.3f}'.format(metrics.accuracy_score(y_pred, Y_test)))
```

```python
test['Prediction'] = y_pred
```
